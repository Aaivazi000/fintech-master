```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('PerformanceGraphExport.csv', sep='\s*,\s*',
                           header=0, encoding='ascii', engine='python')
```

```python
fig, ax = plt.subplots()
```

```python
data.head()
```

```python
data = data.rolling(window=3, center=True, win_type='triang').mean()
```

```python
data
```

```python
pd.to_datetime(data['Date'])
type(data['Date'])
```

```python
pd.to_datetime(data['Date'])
```

```python
data['year'] = pd.to_datetime(data['Date']).dt.year
```

```python
data['doy'] = pd.to_datetime(data['Date']).dt.dayofyear
```

```python

chart = pd.pivot_table(data, index=['doy'],columns=['year'], values=['S&P 500']).plot()
plt.ylabel('Index value');
plt.xlabel('Day of the year');
plt.show()
```
