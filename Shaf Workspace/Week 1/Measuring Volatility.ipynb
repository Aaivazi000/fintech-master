```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
raw_data = pd.read_csv('Information Technology.csv', sep='\s*,\s*',
                           header=0, encoding='ascii', engine='python')
```

```python
raw_data.head(10)
```

```python
raw_data['date'] = pd.to_datetime(raw_data['date'])
```

```python
data = raw_data[(raw_data['date'] >= '2015-01-02') & (raw_data['date'] <= '2015-12-31')]
```

We wish to look the entire IT sector of the S&P 500 in 2015 to determine this sector's volatility for this year. 

```python


#Get the number of unique companies in the sector
company_count = data.Name.nunique()
```

# If we want to look at the volatility of the entire IT sector for 1 year, we need to first add up all of the closing prices for each company for every day, then divide by the number of companies which existed for that year.

```python
from datetime import datetime
from dateutil.parser import parse

avg_closing = [] 

for row in data['date']:
    current_date = row
    total = 0 
    for row2 in data.itertuples():
       if row2.date == current_date: 
            total = (total + row2.close)
    avg = total/company_count
    #print(current_date, avg)
    avg_closing.append({ 'Date':current_date, 'Total':avg })
    
    str_time = current_date.strftime('%Y-%m-%d')
    loop_date = datetime.strptime(str_time, '%Y-%m-%d')
    last_date = datetime.strptime('2015-12-31', '%Y-%m-%d')
    
    if loop_date == last_date:
       # print('current date IS DECEMBER')
        break 


avg_closing_data = pd.DataFrame(avg_closing)        
```

```python
#This is our final list of the average closing prices for each day 
avg_closing_data
```

```python
avg_closing_data['Total'].values.std(ddof=1)
#Standard deviation is the most common method of determining volatility. The STD for 2015 for the IT sector is 3.71
```
