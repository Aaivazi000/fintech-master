```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Import the csv
data = pd.read_csv("all_stocks_5yr.csv",index_col = 'date', parse_dates=["date"])
```

```python
data.tail()
```

```python
data[data.Name == 'AMZN'].close.plot()
```

```python
data[data.Name == 'AMZN'].volume.plot()
```

```python
#MEAN closing prices for each company
data.groupby('Name').close.mean()
```

```python
#MEDIAN closing prices for each company
data.groupby('Name', as_index=False).close.median()
```

```python
#MIN closing prices for each company for the year 2013
data["01/01/2013":"12/31/2013"].sort_values('close').groupby('Name', as_index=False).close.first()
```

```python
#MAX closing prices for each company for the year 2013
data["01/01/2013":"12/31/2013"].sort_values('close').groupby('Name', as_index=False).close.last()
```

```python
list(data['2016'].groupby('Name').close)
data['2016'].groupby('Name').close.plot()
```

```python
data.close.pct_change().plot(figsize=(5, 5), yticks = np.arange(0.0, 5.0, 1))
```

```python
companies = data.Name.unique()
```

```python
#First 5 companies opening price for 2013 
for x in companies[0:5]:
    plt.figure()
    data[data['Name'] ==  x].open.plot(label=x)
    plt.legend(loc='best')
    plt.ylabel("Open Price")
    plt.show()
```

```python
#Percent change of S&P Data 
sp_close_price = data.close
sp_open_price = data.open
((sp_close_price - sp_open_price)).plot(label = 'S&P', figsize=(10,10))
plt.legend(loc='best')
plt.ylabel('Percent change')
plt.xlabel('Date')
```
