## If we want to take a more long-term approach to assessing the performance of a strategy, we can use rolling returns to determine a percentage-based return on any given stock after X years. 

```python
raw_data
```

```python
apple = raw_data[raw_data['Name'] == 'AAPL']
```

```python
apple
```

```python
apple['date'] = pd.to_datetime(apple['date'])

data_2013 = apple[apple['date'].dt.year == 2013]
data_2014 = apple[apple['date'].dt.year == 2014]
data_2015 = apple[apple['date'].dt.year == 2015]
data_2016 = apple[apple['date'].dt.year == 2016]
data_2017 = apple[apple['date'].dt.year == 2017]
```

```python
diff_2013 = (data_2013.iloc[-1].close - data_2013.iloc[0].close) / data_2013.iloc[0].close
diff_2014 = (data_2014.iloc[-1].close - data_2014.iloc[0].close) / data_2014.iloc[0].close
diff_2015 = (data_2015.iloc[-1].close - data_2015.iloc[0].close) / data_2015.iloc[0].close
diff_2016 = (data_2016.iloc[-1].close - data_2016.iloc[0].close) / data_2016.iloc[0].close
diff_2017 = (data_2017.iloc[-1].close - data_2017.iloc[0].close) / data_2017.iloc[0].close
five_year_return = (diff_2013 + diff_2014 + diff_2015 + diff_2016 + diff_2017) / 5 * 100
print(diff_2013 * 100)
print(diff_2014 * 100)
print(diff_2015 * 100)
print(diff_2016 * 100)
print(diff_2017 * 100)
print('Five-year return for Apple:',five_year_return, '%')
```
